```python
from openalgo import api

client = api(api_key="your-openalgo-apikey", host="http://127.0.0.1:5000")


response = client.history(
    symbol="NIFTY",
    exchange="NSE_INDEX",
    interval="1m",
    start_date="2025-10-01",
    end_date="2025-10-10",
)
response
```
